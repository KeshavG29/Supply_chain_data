```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("data/supply_chain_data.csv")
df = pd.DataFrame(data)
print(df)


```

```python
print(df.describe())
```

**Revenue Generated By Products**



> Company generates most of it revenue from the skincare products.


> Revenue generated:


*   **Skincare**  : 2.41628 lakhs
*  **Haircare** : 1.74455 lakhs
*   **Cosmetics** : 1.61521 lakhs














```python
grouped = df.groupby("Product type")["Revenue generated"].sum()
print(grouped)

grouped.plot.bar()
plt.xlabel("Products")
plt.ylabel("Revenue Generated")

plt.show()
```

**Company is having almost 45% of its sale from skincare product.**

```python
sale = df.groupby("Product type")["Number of products sold"].sum()
print(sale)

plt.pie(sale.values , labels = sale.index , autopct = "%0.2f")
plt.show()
```

#          **Finding The Relation Between The Revenue And Price**

```python
sns.lmplot(data = df, x= "Price", y="Revenue generated", hue = "Product type")
# plt.legend(loc='lower right')
```

# **Observations:**


*   Revenue of the skincare product increases with the increase in Price of product.

*   Revenue of the Cosmetics decreases with the increase in Price of product.


*   There is a slight decrease in the Revenue generated of the haircare product with increase in Price.

# **Inference**



> Customers are ready to buy the Skincare products at higher Prices also.







```python
Shiping_revenue = df.groupby("Shipping carriers")['Revenue generated'].sum()
# print(Shiping_revenue)

Shiping_revenue.plot.bar()
```
